```python
#using our previous real estate data 

import pandas as pd 
import matplotlib.pyplot as plt

ct_re_2 = pd.read_csv("Real_Estate_Sales_2001-2020_GL.csv", low_memory=False)
ct_re_2
```

```python
df = ct_re_2.groupby("Property Type")[["Sale Amount"]].mean().round()
df
```

```python
#see all the colors you can choose from at
#https://matplotlib.org/stable/gallery/color/named_colors.html#sphx-glr-gallery-color-named-colors-py


colors = ["orchid", "navy", "limegreen", "coral", "slategray", "pink", "black", "teal", "indigo",
          "cyan", "gold"]

ax = df.plot(kind = "pie", y="Sale Amount", title="Types of Homes Sold",
            ylabel = "", figsize=(10,10), radius=1, startangle=50, labeldistance=1.05,
            colors=colors)
ax.legend(bbox_to_anchor=(.0, .5))

plt.show()
```

```python
#let's use a new data set 

file_name= "Broome_County_Annual_Employee_Earnings__Beginning_2009.csv"

broome_data = pd.read_csv(f"{file_name}")
broome_data #this variable stores the data frame 
```

```python
#we can plot the dataframe with no other arguments, 
#but it doesn't turn out very nicely, does it?

broome_data.plot()
plt.show()
```

```python
#example of customizing a graph
#ax refers to Axes, that is an object from matplotlib

#taking two columns that it makes sense to graph together


ax = broome_data.plot(kind="hist", x="Earnings Year", y="Total Earnings", xlim=(0, 150000), 
                title="Total Earnings Broome County Employees, 2009-2021",
                 figsize = (10,10), fontsize=10, color="Green", legend=False)

ax.set_xlabel("Salary")
ax.set_ylabel("Number of employees")
plt.show()
```
